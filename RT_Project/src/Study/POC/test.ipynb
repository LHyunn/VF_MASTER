```python
from glob import glob
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image_list = glob('/home/VirtualFlaw/Data/IMAGE/Raw_jpg/**/*.jpg', recursive=True)
sd_list = []
for i in image_list:
    if 'SD' in i:
        sd_list.append(i)

print(len(sd_list))
image = np.random.choice(sd_list, 1)
image = cv2.imread(image[0])
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)
plt.imshow(image, cmap='gray')
```
